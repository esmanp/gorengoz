```python
# Use a pipeline as a high-level helper
from transformers import pipeline
```

```python
ner_pipeline = pipeline("ner", model="Tirendaz/roberta-base-NER")
```

```python
yorum = "Fiber 100mb SuperOnline kullanıcısıyım yaklaşık 2 haftadır @Twitch @Kick_Turkey gibi canlı yayın platformlarında 360p yayın izlerken donmalar yaşıyoruz. Başka hiç bir operatörler bu sorunu yaşamazken ben parasını verip alamadığım hizmeti neden ödeyeyim ? @Turkcell "
```

```python
ner_pipeline(yorum)
```

```python
ner_pipeline(yorum, aggregation_strategy = "simple")
```

```python
pipe = pipeline("token-classification", model="akdeniz27/bert-base-turkish-cased-ner")
```

```python
pipe(yorum)
```

```python
pipe(yorum, aggregation_strategy = "simple")
```

```python
pipe2(yorum)
```

```python
pipe2(yorum, aggregation_strategy = "simple")
```
